### * **Test statistic**
### * **Test type (T-test, Z-test, F-test, ANNOVA, Chi-Square, PCA)**

### **The decision rule is: if the p-value for the test is less than 0.05, we reject the null hypothesis, but if it is greater than or equal to 0.05, we fail to reject the null hypothesis.**

### **T-test**
* **Is there a significant difference in the mean age of individuals who tested positive for COVID-19 compared to those who tested negative**

```python
import scipy.stats as stats

# Separate positive and negative test results
positive_cases = df[df['resultPCR'] == 1]
negative_cases = df[df['resultPCR'] == 0]

# Perform two-sample t-test to compare mean ages
t_statistic, p_value = stats.ttest_ind(positive_cases['Age'], negative_cases['Age'])
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Interpret the results

if p_value < 0.05:
    print("There is a statistically significant difference in the mean age of individuals who tested positive for COVID-19 compared to those who tested negative.")
else:
    print("There is no statistically significant difference in the mean age of individuals who tested positive for COVID-19 compared to those who tested negative.")
```

### **Chi-square Test for Smoking and Intubation**
* Is there an association between smoking status and the need for intubation

```python
from scipy.stats import chi2_contingency

# Create contingency table of smoking and intubation
smoking_intubation_table = pd.crosstab(df['Smoking'], df['Intubation'])

chi2_statistic, p_value, _, _ = chi2_contingency(smoking_intubation_table)
print(f"P-value for chi-square test: {p_value}")
```

##### **Interpretation**
**The p-value is less than a chosen significance level (e.g., 0.0139), it indicates that there's evidence to reject the null hypothesis. Thereis no  association between smoking status and the need for intubation.**

### **proportions_ztest**
*  Is there a significant difference in the proportion of COVID-19 positive cases between genders?

```python
from statsmodels.stats.proportion import proportions_ztest

# Create contingency table of gender vs. positive COVID-19 test results
gender_covid_table = pd.crosstab(df['gender'], df['resultPCR'] == 1)

# Perform z-test for proportions
count = gender_covid_table.iloc[:, 1]  # Number of positive cases by gender
nobs = gender_covid_table.sum(axis=1)  # Total cases by gender

z_statistic, p_value = proportions_ztest(count, nobs)
print("zstatistic:",z_statistic)
print(f"P-value for proportions z-test: {p_value}")
```

#### **Interpretation**
* The p-value is less than a chosen significance level (e.g., 0.05)p_value=9.602467021536645e-10, it indicates that there's evidence to reject the null hypothesis. A significant p-value suggests a difference in the proportion of positive COVID-19 cases between genders.
